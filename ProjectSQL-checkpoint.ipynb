Gender Ratio    

```python
human_resource_sql="""
WITH DepartmentCTE AS(
SELECT A.DepartmentID,
       A.Name AS DepartmentName,
       B.BusinessEntityID,
       C.JobTitle,
       C.Gender
FROM HumanResources.Department AS A 
LEFT JOIN  HumanResources.EmployeeDepartmentHistory AS B
ON A.DepartmentID = B.DepartmentID
LEFT JOIN HumanResources.Employee AS C
ON B.BusinessEntityID=C.BusinessEntityID)


SELECT DepartmentName,
       COUNT(JobTitle) AS 'No.ofEmployees',
       SUM(CASE WHEN Gender='M' THEN 1 ELSE 0 END) AS 'Males',
       SUM(CASE WHEN Gender='F' THEN 1 ELSE 0 END) AS 'Females',
       (SUM(CASE WHEN Gender='M' THEN 1 ELSE 0 END)*10)/NULLIF(SUM(CASE WHEN Gender='F' THEN 1 ELSE 0 END),0) AS 'MaleToFemaleper10'
FROM DepartmentCTE
GROUP BY DepartmentName
ORDER BY COUNT(JobTitle) DESC;
"""
hr_df=pd.read_sql(sql=human_resource_sql,con=connection)
```

```python
print(hr_df)
gender_df = hr_df[['DepartmentName', 'Males', 'Females']]
gender_df.set_index('DepartmentName', inplace=True)
ax = gender_df.plot(kind='bar', stacked=True, colormap='tab20b', figsize=(5, 5))
ax.tick_params(axis='x', labelsize=8) 
ax.set_ylabel('Number of Employees',fontsize=9)
ax.set_xlabel('Department',fontsize=9)
ax.set_title('Gender Distribution Department-wise')
plt.legend(title='Gender', loc='upper right', labels=['Male', 'Female'])
plt.show()
```

Number of products: components used in manufacturing and products for sale.

```python
nbrofproducts_sql="""
WITH LocationCTE AS (
SELECT A.LocationID,
       A.Name AS 'StorageLocation',
       B.ProductID,
       B.Shelf,
       B.Bin,
       C.Name AS 'Product/Part Name',
       (CASE WHEN C.FinishedGoodsFlag='0' THEN 'Part-notsold' ELSE 'Product-sold' END) AS 'Product/Part',
       C.FinishedGoodsFlag
FROM Production.Location AS A 
     LEFT JOIN Production.ProductInventory AS B
ON A.LocationID = B.LocationID LEFT JOIN Production.Product AS C ON B.ProductID=C.ProductID )


SELECT StorageLocation,
       COUNT('Product/Part') AS NbrofProducts, 
       SUM(CASE WHEN FinishedGoodsFlag='0' THEN 1 ELSE 0 END) AS Components,
       SUM(CASE WHEN FinishedGoodsFlag='1' THEN 1 ELSE 0 END) AS ProductsforSale
FROM LocationCTE
GROUP BY StorageLocation
ORDER BY COUNT('Product/Part') DESC;
"""
products_df=pd.read_sql(sql=nbrofproducts_sql,con=connection)
```

```python
print(products_df)

products_df.set_index('StorageLocation',inplace=True)

ax = products_df[['Components','ProductsforSale']].plot(kind='bar', stacked=True, colormap='tab10', figsize=(5, 5))
ax.tick_params(axis='x', labelsize=8) 
ax.set_ylabel('Number of products',fontsize=9)
ax.set_xlabel('Storage Location',fontsize=9)
ax.set_title('Number of products and their locations')
plt.legend(title='Type', loc='upper right', labels=['Components', 'ProductsforSale'])
plt.show()
```

Stock levels

```python
stocklevels_sql="""
WITH StockLevelCTE AS (
SELECT A.LocationID,
       A.Name AS StorageLocation,
       B.ProductID,
       B.Shelf,
       B.Bin,
       B.Quantity,
       C.Name AS ProductorPartName ,
       C.SafetyStockLevel,
       C.ReorderPoint,
       C.DaysToManufacture,
       C.FinishedGoodsFlag
FROM Production.Location AS A 
     LEFT JOIN Production.ProductInventory AS B
ON A.LocationID = B.LocationID LEFT JOIN Production.Product AS C ON B.ProductID=C.ProductID )

SELECT StorageLocation,
       ProductorPartName,
       Shelf,
       Bin,
       Quantity,
       SafetyStockLevel,
       ReorderPoint,
       (CASE WHEN Quantity>=SafetyStockLevel THEN 'High' 
       WHEN Quantity<SafetyStockLevel AND Quantity>ReorderPoint THEN 'Medium'
       ELSE 'Low' END) AS 'QuantityLevels',
       (CASE WHEN FinishedGoodsFlag='0' THEN 'Part' ELSE 'Product' END) AS ProductOrPart
FROM StockLevelCTE
"""
stocklevels_df=pd.read_sql(sql=stocklevels_sql,con=connection)
```

```python
print(stocklevels_df)
quantitylevel_counts = stocklevels_df['QuantityLevels'].value_counts()
fig, axs = plt.subplots(figsize=(5, 5))
axs.pie(quantitylevel_counts, labels=quantitylevel_counts.index,autopct='%1.1f%%',startangle=70, colors=plt.cm.Paired.colors)
axs.set_title('Quantity Levels')
plt.show()
```

Scrap information

```python
scrap_sql="""
WITH ScrapCTE AS(
SELECT A.ProductID,
       A.ScrappedQty,
       A.ScrapReasonID,
       B.Name AS ScrapReason
FROM Production.WorkOrder AS A LEFT JOIN Production.ScrapReason AS B
ON A.ScrapReasonID=B.ScrapReasonID
WHERE A.ScrappedQty<>0)


SELECT TOP 10 SUM(ScrappedQty) AS TotalScrappedQty,
       ScrapReason
FROM ScrapCTE
GROUP BY ScrapReason
ORDER BY TotalScrappedQty DESC;
"""
scrap_df=pd.read_sql(sql=scrap_sql,con=connection)
```

```python
print(scrap_df)

fig, axs = plt.subplots(figsize=(5, 5))
axs.pie(scrap_df['TotalScrappedQty'],labels=scrap_df['ScrapReason'],autopct='%1.1f%%', colors=plt.cm.Paired.colors)
axs.set_title('Top 10 reasons for scrap')
plt.show()
```

Confidence interval for difference between the proportions of total sales in May and October:

```python
import math
import numpy as np
from scipy import stats
```

```python
#extracting data for second highest and third highest selling months 
may_sales = salespermonth_df[salespermonth_df['Month'] == 5]['Sale'].sum()
october_sales = salespermonth_df[salespermonth_df['Month'] == 10]['Sale'].sum()
total_sales = salespermonth_df['Sale'].sum()  # Summing all months
#calculating proportions
p1_may=may_sales/total_sales
p2_october=october_sales/total_sales
#confidence interval for difference between the quantity of sales in May and October
std_error_diff = math.sqrt((p1_may * (1 - p1_may) / may_sales) + (p2_october * (1 - p2_october) / october_sales))
z_score = 1.96
lower_bound = ( p1_may-p2_october) - z_score * std_error_diff
upper_bound = ( p1_may-p2_october) + z_score * std_error_diff
#confidence intervals
print('Confidence interval upper bound= ',upper_bound)
print('Confidence interval lower bound= ',lower_bound)
```

```python

may_sales = salespermonth_df[salespermonth_df['Month'] == 5]['Sale'].sum()
october_sales = salespermonth_df[salespermonth_df['Month'] == 10]['Sale'].sum()
total_sales = salespermonth_df['Sale'].sum()  # Summing all months

# calculating proportions
p1_may = may_sales / total_sales
p2_october = october_sales / total_sales

# calculating pooled proportion
p_combined = (may_sales * p1_may + october_sales * p2_october) / (may_sales + october_sales)

# calculating standard error
std_error_diff = np.sqrt((p_combined * (1 - p_combined) / may_sales) + (p_combined * (1 - p_combined) / october_sales))

# calculating z-score
z_score = (p1_may-p2_october) / std_error_diff

# calculating p-value
p_value = 1 - stats.norm.cdf(z_score)

# hypothesis testing
percentage = 0.05
if p_value < percentage:
    print("Reject the null hypothesis. There is a significant difference in proportions of sales in May and October.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in proportions of sales in May and October.")
```
